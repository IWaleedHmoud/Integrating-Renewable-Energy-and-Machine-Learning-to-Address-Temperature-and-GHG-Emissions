```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Excel file
df = pd.read_excel('PCC Data.xlsx')

# Keep only the needed columns
selected_cols = df[[9, 179, 20]]  # column names are numbers

# Rename columns
selected_cols.columns = ["CO2", "Radiation", "Temperature"]

# Compute Pearson correlation matrix
corr_matrix = selected_cols.corr(method='pearson')

# Plot heatmap with new labels
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1,
            xticklabels=["CO2", "Radiation", "Temperature"],
            yticklabels=["CO2", "Radiation", "Temperature"])
plt.title('Pearson Correlation Coefficient (PCC) Matrix')
plt.show()
```
